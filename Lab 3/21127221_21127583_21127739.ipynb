### 4.1. Ph√¢n t√≠ch t·ª∑ l·ªá tr√πng l·∫∑p (duplicate) v√† x·ª≠ l√Ω c√°c d√≤ng b·ªã tr√πng l·∫∑p (n·∫øu c·∫ßn)

Ta s·ª≠ d·ª•ng ph∆∞∆°ng th·ª©c "duplicated()" c·ªßa ƒë·ªëi t∆∞·ª£ng DataFrame ƒë·ªÉ ki·ªÉm tra xem c√≥ d√≤ng n√†o xu·∫•t hi·ªán nhi·ªÅu h∆°n m·ªôt l·∫ßn hay kh√¥ng.

```python
# Reference: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.duplicated.html
n_duplicated_rows = raw_df.duplicated(keep='first').sum()
print(f"B·ªô d·ªØ li·ªáu th√¥ c√≥ {n_duplicated_rows} d√≤ng b·ªã tr√πng l·∫∑p.")
```

Nh·∫≠n x√©t:

- Xem qua b·∫£ng d·ªØ li·ªáu c·ªßa file CSV v√† quan s√°t k·∫øt qu·∫£ ph√¢n t√≠ch th√¨ c√≥ v·∫ª nh∆∞ kh√¥ng c√≥ m·ªôt b√¥ng hoa n√†o xu·∫•t hi·ªán nhi·ªÅu h∆°n m·ªôt l·∫ßn trong b·ªô d·ªØ li·ªáu. Nghƒ©a l√† b·ªô d·ªØ li·ªáu th√¥ kh√¥ng c√≥ d√≤ng n√†o b·ªã tr√πng l·∫∑p, v√† ta c√≥ th·ªÉ chuy·ªÉn sang c√°c b∆∞·ªõc ph√¢n t√≠ch ti·∫øp theo.

### 4.2. Ph√¢n t√≠ch t·ª∑ l·ªá thi·∫øu gi√° tr·ªã ·ªü m·ªói c·ªôt (missing rate) v√† x·ª≠ l√Ω c√°c c·ªôt c√≥ gi√° tr·ªã thi·∫øu (n·∫øu c·∫ßn)

Ta s·ª≠ d·ª•ng ph∆∞∆°ng th·ª©c "isnull()" c·ªßa ƒë·ªëi t∆∞·ª£ng DataFrame ƒë·ªÉ ki·ªÉm tra xem c√≥ √¥ n√†o trong b·∫£ng d·ªØ li·ªáu b·ªã thi·∫øu gi√° tr·ªã hay kh√¥ng. T·ª´ ƒë√≥ ta suy ra t·ª∑ l·ªá thi·∫øu gi√° tr·ªã ·ªü m·ªói c·ªôt.

```python
# ƒê·∫øm s·ªë l∆∞·ª£ng gi√° tr·ªã b·ªã thi·∫øu trong m·ªói c·ªôt
missing_values = raw_df.isnull().sum()

# S·∫Øp x·∫øp c√°c d√≤ng gi·∫£m d·∫ßn theo s·ªë l∆∞·ª£ng gi√° tr·ªã b·ªã thi·∫øu
missing_values = missing_values.sort_values(ascending=False)

# Chuy·ªÉn Series v·ªÅ DataFrame ƒë·ªÉ d·ªÖ quan s√°t
missing_values = missing_values.reset_index()
missing_values.columns = ["T√™n c·ªôt", "S·ªë l∆∞·ª£ng gi√° tr·ªã b·ªã thi·∫øu"]

# T·∫°o c·ªôt m·ªõi cho bi·∫øt t·ª∑ l·ªá thi·∫øu gi√° tr·ªã c·ªßa m·ªói c·ªôt
missing_values["T·ª∑ l·ªá thi·∫øu gi√° tr·ªã (%)"] = \
    missing_values["S·ªë l∆∞·ª£ng gi√° tr·ªã b·ªã thi·∫øu"] / len(raw_df) * 100

# C·∫≠p nh·∫≠t ch·ªâ m·ª•c c·ªßa DataFrame
missing_values.index += 1
missing_values.index.name = "STT"

# Hi·ªÉn th·ªã k·∫øt qu·∫£
missing_values
```

Nh·∫≠n x√©t:

- T·ª´ k·∫øt qu·∫£ ph√¢n t√≠ch d·ªØ li·ªáu, ta th·∫•y r·∫±ng kh√¥ng c√≥ c·ªôt n√†o trong b·ªô d·ªØ li·ªáu th√¥ b·ªã thi·∫øu gi√° tr·ªã. ƒê√¢y l√† m·ªôt ƒëi·ªÅu r·∫•t ƒë√°ng m·ª´ng, cho th·∫•y qu√° tr√¨nh thu th·∫≠p d·ªØ li·ªáu ƒë∆∞·ª£c ti·∫øn h√†nh r·∫•t t·ªët. Nh∆∞ v·∫≠y, ta kh√¥ng c·∫ßn ph·∫£i x·ª≠ l√Ω g√¨ th√™m ·ªü b∆∞·ªõc n√†y v√† c√≥ th·ªÉ chuy·ªÉn sang c√°c b∆∞·ªõc ph√¢n t√≠ch ti·∫øp theo.

### 4.3. Lo·∫°i b·ªè c√°c c·ªôt kh√¥ng c√≥ nhi·ªÅu √Ω nghƒ©a cho vi·ªác ph√¢n t√≠ch d·ªØ li·ªáu theo m·ª•c ti√™u ƒë√£ ƒë·ªÅ ra

Theo b·∫£ng m√¥ t·∫£ v·ªÅ c√°c c·ªôt m√† ta ƒë√£ vi·∫øt ·ªü tr√™n, c·ªôt "Id" cho bi·∫øt m√£ ƒë·ªãnh danh c·ªßa m·ªói m·∫´u d·ªØ li·ªáu, v√† m·ªói m·∫´u d·ªØ li·ªáu s·∫Ω c√≥ m·ªôt m√£ ƒë·ªãnh danh ri√™ng. ƒê√¢y l√† th√¥ng tin kh√¥ng c√≥ nhi·ªÅu √Ω nghƒ©a trong vi·ªác gi√∫p ta hi·ªÉu r√µ h∆°n c√°c ƒë·∫∑c ƒëi·ªÉm n·ªïi b·∫≠t c·ªßa m·ªói lo√†i hoa di√™n vƒ©, ƒë·ªÉ t·ª´ ƒë√≥ h·ªó tr·ª£ cho b√†i to√°n ph√¢n l·ªõp c√°c lo√†i hoa di√™n vƒ© kh√°c nhau.

Nh∆∞ v·∫≠y, ta quy·∫øt ƒë·ªãnh s·∫Ω lo·∫°i b·ªè c·ªôt "Id" kh·ªèi b·ªô d·ªØ li·ªáu v√† ch·ªâ gi·ªØ l·∫°i c√°c ƒë·∫∑c tr∆∞ng quan tr·ªçng (l√† c√°c c·ªôt c√≤n l·∫°i).

```python
# Lo·∫°i b·ªè c·ªôt "Id" kh·ªèi DataFrame
raw_df = raw_df.drop("Id", axis=1)
```

ƒê·∫øm s·ªë d√≤ng v√† s·ªë c·ªôt c√≤n l·∫°i c·ªßa b·ªô d·ªØ li·ªáu

### 4.4. Ph√¢n t√≠ch ki·ªÉu d·ªØ li·ªáu c·ªßa m·ªói c·ªôt v√† x·ª≠ l√Ω c√°c c·ªôt c√≥ ki·ªÉu d·ªØ li·ªáu ch∆∞a ph√π h·ª£p (n·∫øu c·∫ßn)

Thay v√¨ s·ª≠ d·ª•ng thu·ªôc t√≠nh "dtypes" c·ªßa ƒë·ªëi t∆∞·ª£ng DataFrame, ta c√≥ th·ªÉ ph√¢n t√≠ch d·ªØ li·ªáu trong m·ªói c·ªôt ƒë·ªÉ x√°c ƒë·ªãnh ki·ªÉu d·ªØ li·ªáu c·ªßa c·ªôt t∆∞∆°ng ·ª©ng.

```python
# H√†m h·ªó tr·ª£ r√∫t tr√≠ch ki·ªÉu d·ªØ li·ªáu c·ªßa m·ªói c·ªôt
def _extract_data_type(s: str) -> str:
    """ Extract the actual data type from the string 
    representing the data type of column

    Examples:
    [1]: s = "<class 'numpy.int64'>"
    >>> return "int64"
    
    [2]: s = "<class 'str'>"
    >>> return "str"
    """

    s = s.split("\'")[-2]
    s = s.split(".")[-1]
    return s

# R√∫t tr√≠ch ki·ªÉu d·ªØ li·ªáu c·ªßa m·ªói c·ªôt v√† l∆∞u v√†o Series
dtypes = raw_df.iloc[0].apply(
    lambda x: _extract_data_type(str(type(x)))
)

# Chuy·ªÉn Series v·ªÅ DataFrame ƒë·ªÉ d·ªÖ quan s√°t
dtypes = dtypes.reset_index()
dtypes.columns = ["T√™n c·ªôt", "Ki·ªÉu d·ªØ li·ªáu"]

# C·∫≠p nh·∫≠t ch·ªâ m·ª•c c·ªßa DataFrame
dtypes.index += 1
dtypes.index.name = "STT"

# Hi·ªÉn th·ªã k·∫øt qu·∫£
dtypes
```

Nh·∫≠n x√©t:

  - C·ªôt "Species" c√≥ ki·ªÉu d·ªØ li·ªáu "str" cho bi·∫øt t√™n c·ªßa lo√†i hoa m√† m·∫´u d·ªØ li·ªáu thu·ªôc v·ªÅ.

### 4.5. Ph√¢n t√≠ch ph√¢n b·ªë c·ªßa c√°c gi√° tr·ªã trong m·ªói c·ªôt c√≥ ki·ªÉu d·ªØ li·ªáu d·∫°ng s·ªë (numerical) v√† x·ª≠ l√Ω c√°c c·ªôt c√≥ gi√° tr·ªã b·∫•t th∆∞·ªùng

V·ªõi m·ªói c·ªôt c√≥ ki·ªÉu d·ªØ li·ªáu d·∫°ng s·ªë, ta s·∫Ω t√≠nh:
- T·ª∑ l·ªá thi·∫øu gi√° tr·ªã (t·ª´ 0 ƒë·∫øn 100).
- Gi√° tr·ªã t·ªëi thi·ªÉu.
- Gi√° tr·ªã t·ª© ph√¢n v·ªã th·ª© nh·∫•t.
- Gi√° tr·ªã t·ª© ph√¢n v·ªã th·ª© hai (gi√° tr·ªã trung v·ªã).
- Gi√° tr·ªã t·ª© ph√¢n v·ªã th·ª© ba.
- Gi√° tr·ªã t·ªëi ƒëa.

```python
# M·ªôt s·ªë h√†m ti·ªán √≠ch
def missing_ratio(s):
    return (s.isna().mean() * 100).round(1)

def median(df):
    return (df.quantile(0.5)).round(1)

def lower_quartile(df):
    return (df.quantile(0.25)).round(1)

def upper_quartile(df):
    return (df.quantile(0.75)).round(1)

# T·∫°o m·ªôt DataFrame ch·ªâ ch·ª©a c√°c thu·ªôc t√≠nh s·ªë
num_col_df = raw_df.select_dtypes(include=[np.number])

# T·∫°o b·∫£ng m√¥ t·∫£ cho c√°c thu·ªôc t√≠nh s·ªë
num_col_info_df = num_col_df.agg([
    missing_ratio, "min", lower_quartile,
    median, upper_quartile, "max"
])

# Ch·ªânh s·ª≠a DataFrame ƒë·ªÉ hi·ªÉn th·ªã k·∫øt qu·∫£ t·ªët h∆°n
## ƒê·ªïi t√™n c√°c d√≤ng (c√°c ch·ªâ m·ª•c)
num_col_info_df.index = [
    "T·ª∑ l·ªá thi·∫øu gi√° tr·ªã (%)", "Gi√° tr·ªã t·ªëi thi·ªÉu", 
    "T·ª© ph√¢n v·ªã th·ª© nh·∫•t (Q1)", "Gi√° tr·ªã trung v·ªã (Q2)", 
    "T·ª© ph√¢n v·ªã th·ª© ba (Q3)", "Gi√° tr·ªã t·ªëi ƒëa"
]
## ƒê·∫∑t t√™n cho c√°c ch·ªâ m·ª•c v√† c√°c c·ªôt
num_col_info_df.index.name = "Ch·ªâ s·ªë th·ªëng k√™"
num_col_info_df.columns.name = "T√™n c·ªôt"

# Hi·ªÉn th·ªã k·∫øt qu·∫£
num_col_info_df
```

### 4.6. Ph√¢n t√≠ch ph√¢n b·ªë c·ªßa c√°c gi√° tr·ªã trong m·ªói c·ªôt c√≥ ki·ªÉu d·ªØ li·ªáu d·∫°ng ph√¢n lo·∫°i (categorical) v√† x·ª≠ l√Ω c√°c c·ªôt c√≥ gi√° tr·ªã b·∫•t th∆∞·ªùng

V·ªõi m·ªói c·ªôt c√≥ ki·ªÉu d·ªØ li·ªáu d·∫°ng ph√¢n lo·∫°i, ta s·∫Ω t√≠nh:
- T·ª∑ l·ªá thi·∫øu gi√° tr·ªã (t·ª´ 0 ƒë·∫øn 100).
- S·ªë l∆∞·ª£ng c√°c gi√° tr·ªã kh√°c nhau.
- T·ª∑ l·ªá xu·∫•t hi·ªán (t·ª´ 0 ƒë·∫øn 100) c·ªßa m·ªói gi√° tr·ªã.

```python
# M·ªôt s·ªë h√†m ti·ªán √≠ch
def missing_ratio(s):
    return (s.isna().mean() * 100).round(1)

def num_values(s):
    return len(s.value_counts())

def value_ratios(s):
    # T√≠nh t·ª∑ l·ªá ph·∫ßn trƒÉm c·ªßa m·ªói gi√° tr·ªã
    total_count = (~s.isna()).sum()
    ratios = (s.value_counts() / total_count * 100).round(1)

    # Chuy·ªÉn ki·ªÉu d·ªØ li·ªáu c·ªßa ch·ªâ m·ª•c th√†nh `str`    
    ratios.index = ratios.index.astype(str)
    
    return ratios.to_dict()

# T·∫°o m·ªôt DataFrame ch·ªâ ch·ª©a c√°c thu·ªôc t√≠nh ph√¢n lo·∫°i
cate_col_df = raw_df.select_dtypes(include=['object'])

# T·∫°o b·∫£ng m√¥ t·∫£ cho c√°c thu·ªôc t√≠nh ph√¢n lo·∫°i
cate_col_info_df = cate_col_df.agg([
    missing_ratio, num_values, value_ratios
])
cate_col_info_df = cate_col_info_df.T

# Ch·ªânh s·ª≠a DataFrame ƒë·ªÉ hi·ªÉn th·ªã k·∫øt qu·∫£ t·ªët h∆°n
## ƒê·ªïi t√™n c√°c c·ªôt
cate_col_info_df.columns = [
    "T·ª∑ l·ªá thi·∫øu gi√° tr·ªã (%)", "S·ªë l∆∞·ª£ng gi√° tr·ªã kh√°c nhau",
    "T·ª∑ l·ªá xu·∫•t hi·ªán (%) c·ªßa m·ªói gi√° tr·ªã"
]
## ƒê·∫∑t t√™n cho ch·ªâ m·ª•c
cate_col_info_df.index.name = "T√™n c·ªôt"

# Hi·ªÉn th·ªã k·∫øt qu·∫£
cate_col_info_df
```

### 5.2. Ph√¢n t√≠ch ph√¢n ph·ªëi ƒë·ªëi v·ªõi c√°c bi·∫øn ƒë·ªãnh l∆∞·ª£ng

##### 5.2.1. Ph√¢n t√≠ch ph√¢n ph·ªëi c·ªßa chi·ªÅu d√†i l√° ƒë√†i (SepalLengthCm)

ƒê·∫ßu ti√™n, ta v·∫Ω bi·ªÉu ƒë·ªì th·ªÉ hi·ªán m·∫≠t ƒë·ªô ph√¢n b·ªë c·ªßa c√°c gi√° tr·ªã trong c·ªôt "SepalLengthCm".

```python
# H√†m ti·ªán √≠ch gi√∫p t·∫°o t√™n cho thu·ªôc t√≠nh
def _generate_feature_name(feature: str) -> str:
    """ S·ª≠ d·ª•ng t√™n thu·ªôc t√≠nh ƒë·ªÉ sinh ra t√™n th√≠ch h·ª£p cho bi·ªÉu ƒë·ªì.

    Args:
        feature (str): T√™n thu·ªôc t√≠nh c·∫ßn tr·ª±c quan.

    Returns:
        (str): T√™n th√≠ch h·ª£p cho bi·ªÉu ƒë·ªì.
    """
    
    name = ""
    feature = feature.lower()
    
    if   "length" in feature: name += "Chi·ªÅu d√†i"
    elif "width" in feature : name += "Chi·ªÅu r·ªông"
    
    if   "sepal" in feature: name += " l√° ƒë√†i"
    elif "petal" in feature: name += " c√°nh hoa"
    
    return name

# H√†m ti·ªán √≠ch gi√∫p tr·ª±c quan c√°c thu·ªôc t√≠nh s·ªë
def _plot_distribution(ser: pd.Series):
    """ Tr·ª±c quan ph√¢n ph·ªëi c·ªßa thu·ªôc t√≠nh d·∫°ng s·ªë b·∫±ng bi·ªÉu ƒë·ªì m·∫≠t ƒë·ªô

    Args:
        ser (pd.Series): Series ch·ª©a d·ªØ li·ªáu c·∫ßn tr·ª±c quan.
    """
    
    # T·∫°o bi·ªÉu ƒë·ªì density plot
    plt.figure(figsize=(10, 6))
    sns.kdeplot(ser, color="skyblue", fill=True)

    # V·∫Ω c√°c ƒë∆∞·ªùng th·∫≥ng ƒë·ª©ng cho c√°c gi√° tr·ªã t·ª© ph√¢n v·ªã    
    plt.axvline(ser.quantile(0.25), color="#02367B", linestyle="--", label="Q1")
    plt.axvline(ser.median(), color="#FF4F58", linestyle="--", label="Q2")
    plt.axvline(ser.quantile(0.75), color="#097969", linestyle="--", label="Q3")

    # V·∫Ω c√°c ƒë∆∞·ªùng th·∫≥ng ƒë·ª©ng cho gi√° tr·ªã min v√† max
    plt.axvline(ser.min(), color="#191970", linestyle="-", label="Min")
    plt.axvline(ser.max(), color="#023020", linestyle="-", label="Max")
    
    # V·∫Ω kho·∫£ng tin c·∫≠y 95% cho gi√° tr·ªã trung b√¨nh
    err = 1.96 * ser.std() / np.sqrt(len(ser))
    plt.axvspan(ser.mean()-err, ser.mean()+err, alpha=0.2, color="yellow")
    plt.axvline(ser.mean(), color="#FFAA1D", linestyle="-.", label="Mean")

    # ƒê·∫∑t t√™n cho c√°c tr·ª•c v√† ti√™u ƒë·ªÅ
    feature_name = _generate_feature_name(ser.name)
    plt.title(f"Bi·ªÉu ƒë·ªì m·∫≠t ƒë·ªô c·ªßa `{feature_name}`")
    plt.xlabel(f"{feature_name} (cm)")
    plt.ylabel("M·∫≠t ƒë·ªô")
    plt.legend()
    # print(ser.mean()-err, ser.mean()+err)


# V·∫Ω bi·ªÉu ƒë·ªì m·∫≠t ƒë·ªô cho thu·ªôc t√≠nh "SepalLengthCm"
feature = "SepalLengthCm"
_plot_distribution(raw_df[feature])
plt.show()
```

Sau ƒë√≥, ta t·∫°o m·ªôt b·∫£ng m√¥ t·∫£ c√°c ƒë·∫°i l∆∞·ª£ng th·ªëng k√™ cho c·ªôt "SepalLengthCm".

```python
# H√†m ti·ªán √≠ch gi√∫p t·∫°o b·∫£ng th·ªëng k√™ cho c√°c thu·ªôc t√≠nh d·∫°ng s·ªë
def _generate_description(ser: pd.Series) -> pd.DataFrame:
    """ T·∫°o b·∫£ng m√¥ t·∫£ c√°c gi√° tr·ªã th·ªëng k√™ cho thu·ªôc t√≠nh d·∫°ng s·ªë

    Args:
        ser (pd.Series): Series ch·ª©a d·ªØ li·ªáu c·∫ßn m√¥ t·∫£.

    Returns:
        pd.DataFrame: B·∫£ng m√¥ t·∫£ c√°c gi√° tr·ªã th·ªëng k√™.
    """
    
    # T·∫°o b·∫£ng m√¥ t·∫£ c√°c gi√° tr·ªã th·ªëng k√™
    description = ser.agg(
        ["min", lower_quartile, median, upper_quartile, "max", 
         "mean", "std", pd.Series.skew, pd.Series.kurtosis]
    ).to_frame()

    # ƒê·∫∑t l·∫°i t√™n cho c·ªôt v√† ch·ªâ m·ª•c
    feature_name = _generate_feature_name(ser.name)
    description.columns = ["Gi√° tr·ªã"]
    description.index.name = f"ƒê·∫°i l∆∞·ª£ng th·ªëng k√™ cho `{feature_name} (cm)`"
    
    return description

# T·∫°o b·∫£ng m√¥ t·∫£ c√°c gi√° tr·ªã th·ªëng k√™ cho thu·ªôc t√≠nh "SepalLengthCm"
_generate_description(raw_df["SepalLengthCm"])
```

  - Tuy gi√° tr·ªã trung b√¨nh c√≥ l·ªõn h∆°n gi√° tr·ªã trung v·ªã m·ªôt ch√∫t nh∆∞ng kh√¥ng ƒë√°ng k·ªÉ.

ƒê·∫ßu ti√™n, ta v·∫Ω bi·ªÉu ƒë·ªì th·ªÉ hi·ªán ph√¢n b·ªë cho c√°c gi√° tr·ªã trong c·ªôt "SepalLengthCm" c·ªßa c√°c m·∫´u d·ªØ li·ªáu thu·ªôc v·ªÅ lo√†i hoa "Iris-setosa".

```python
# H√†m ti·ªán √≠ch gi√∫p tr·ª±c quan bi·ªÉu ƒë·ªì m·∫≠t ƒë·ªô v√† bi·ªÉu ƒë·ªì h·ªôp c·ªßa thu·ªôc t√≠nh s·ªë
def _density_and_box_plot(ser: pd.Series, color: str, species: str):
    """ Tr·ª±c quan ph√¢n ph·ªëi c·ªßa thu·ªôc t√≠nh d·∫°ng s·ªë b·∫±ng bi·ªÉu ƒë·ªì m·∫≠t ƒë·ªô

    Args:
    - ser (pd.Series): Series ch·ª©a d·ªØ li·ªáu c·∫ßn tr·ª±c quan.
    - color (str): M√†u s·∫Øc cho bi·ªÉu ƒë·ªì.
    - species (str): T√™n lo√†i hoa c·∫ßn tr·ª±c quan.
    """
    
    # T·∫°o t√™n cho thu·ªôc t√≠nh
    feature_name = _generate_feature_name(ser.name)

    # T·∫°o 2 subplots
    fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 6), sharex=True)

    # -------------------- V·∫Ω bi·ªÉu ƒë·ªì m·∫≠t ƒë·ªô ·ªü tr√™n --------------------
    sns.kdeplot(ser, color=color, fill=True, ax=axes[0])

    # V·∫Ω c√°c ƒë∆∞·ªùng th·∫≥ng ƒë·ª©ng cho c√°c gi√° tr·ªã t·ª© ph√¢n v·ªã    
    axes[0].axvline(ser.quantile(0.25), color="#02367B", linestyle="--", label="Q1")
    axes[0].axvline(ser.median(), color="#FF4F58", linestyle="--", label="Q2")
    axes[0].axvline(ser.quantile(0.75), color="#097969", linestyle="--", label="Q3")

    # V·∫Ω c√°c ƒë∆∞·ªùng th·∫≥ng ƒë·ª©ng cho gi√° tr·ªã min v√† max
    axes[0].axvline(ser.min(), color="#191970", linestyle="-", label="Min")
    axes[0].axvline(ser.max(), color="#023020", linestyle="-", label="Max")

    # V·∫Ω kho·∫£ng tin c·∫≠y 95% cho gi√° tr·ªã trung b√¨nh
    err = 1.96 * ser.std() / np.sqrt(len(ser))
    axes[0].axvspan(ser.mean()-err, ser.mean()+err, alpha=0.2, color="yellow")
    axes[0].axvline(ser.mean(), color="#FFAA1D", linestyle="-.", label="Mean")

    # ƒê·∫∑t t√™n cho c√°c tr·ª•c, ti√™u ƒë·ªÅ v√† hi·ªÉn th·ªã ch√∫ th√≠ch
    axes[0].legend()
    axes[0].set_ylabel("M·∫≠t ƒë·ªô")
    axes[0].set_title(f"Bi·ªÉu ƒë·ªì th·ªÉ hi·ªán ph√¢n b·ªë `{feature_name}` c·ªßa lo√†i `{species}`")

    # -------------------- V·∫Ω bi·ªÉu ƒë·ªì h·ªôp ·ªü d∆∞·ªõi --------------------
    sns.boxplot(x=ser, color=color, saturation=0.4, #fill=True, 
                linewidth=1.5, ax=axes[1])

    # Xo√° c√°c tr·ª•c y v√† ƒë·∫∑t nh√£n cho tr·ª•c x
    axes[1].set_yticks([])
    axes[1].set_xlabel(f"{feature_name} (cm)")
    # print(ser.mean()-err, ser.mean()+err)


# V·∫Ω bi·ªÉu ƒë·ªì m·∫≠t ƒë·ªô v√† bi·ªÉu ƒë·ªì h·ªôp cho thu·ªôc t√≠nh "SepalLengthCm"
#   c·ªßa lo√†i "Iris-setosa"
species = "Iris-setosa"
feature = "SepalLengthCm"
_density_and_box_plot(ser=setosa_df[feature], color=color_map[species], 
                      species=species)
plt.show()
```

## Ti·ªÅn x·ª≠ l√Ω d·ªØ li·ªáu

·ªû ƒë√¢y, ta kh√¥ng c·∫ßn ti·ªÅn x·ª≠ l√Ω m√† s·∫Ω ƒëi lu√¥n v√†o ph√¢n t√≠ch d·ªØ li·ªáu.

## Ph√¢n t√≠ch d·ªØ li·ªáu

Ta s·∫Ω d√πng bi·ªÉu ƒë·ªì ph√¢n t√°n ƒë·ªÉ th·ªÉ hi·ªán m·ªëi quan h·ªá gi·ªØa c√°c th√¥ng s·ªë k√≠ch th∆∞·ªõc c·ªßa c√°nh hoa theo t·ª´ng lo√†i hoa di√™n vƒ© kh√°c nhau. Th√¥ng qua ƒë√≥, ta s·∫Ω c√≥ th·ªÉ x√°c ƒë·ªãnh xem li·ªáu c√°c lo√†i hoa c√≥ th·ªÉ ƒë∆∞·ª£c ph√¢n t√°ch tuy·∫øn t√≠nh v·ªõi nhau hay kh√¥ng.

```python
# ƒê·ªãnh nghƒ©a 2 thu·ªôc t√≠nh c·∫ßn d√πng
x_feature = "PetalLengthCm"
y_feature = "PetalWidthCm"

# T·∫°o t√™n cho c√°c thu·ªôc t√≠nh
x_feature_name = _generate_feature_name(x_feature)
y_feature_name = _generate_feature_name(y_feature)

# V·∫Ω jointplot
# Reference: https://seaborn.pydata.org/generated/seaborn.jointplot.html
p = sns.jointplot(
    data=raw_df, x=x_feature, y=y_feature, hue="Species",
    palette=color_map, height=8, kind="scatter"
)

# Th√™m ti√™u ƒë·ªÅ cho bi·ªÉu ƒë·ªì
# Reference: https://stackoverflow.com/questions/60358228/how-to-set-title-on-seaborn-jointplot
p.figure.suptitle(f"Bi·ªÉu ƒë·ªì th·ªÉ hi·ªán m·ªëi quan h·ªá gi·ªØa `{x_feature_name}` v√† `{y_feature_name}` theo t·ª´ng lo√†i hoa")
p.figure.subplots_adjust(top=0.95) # Reduce plot to make room 
p.figure.tight_layout()

# Th√™m t√™n cho c√°c tr·ª•c
plt.xlabel(f"{x_feature_name} (cm)")
plt.ylabel(f"{y_feature_name} (cm)")

plt.show()
```

Sau ƒë√≥, ta t·∫°o b·∫£ng m√¥ t·∫£ h·ªá s·ªë t∆∞∆°ng quan Pearson gi·ªØa chi·ªÅu d√†i v√† chi·ªÅu r·ªông c√°nh hoa ·ªü c√°c lo√†i hoa di√™n vƒ© kh√°c nhau.

```python
# T·∫°o b·∫£ng m√¥ t·∫£ h·ªá s·ªë t∆∞∆°ng quan gi·ªØa hai thu·ªôc t√≠nh
description_df = pd.DataFrame(
    data=[
       setosa_df[[x_feature,y_feature]].corr(method='pearson').iloc[1,0],
       versicolor_df[[x_feature,y_feature]].corr(method='pearson').iloc[1,0],
       virginica_df[[x_feature,y_feature]].corr(method='pearson').iloc[1,0],
    ], 
    columns=[f"H·ªá s·ªë t∆∞∆°ng quan Pearson gi·ªØa chi·ªÅu d√†i v√† chi·ªÅu r·ªông c√°nh hoa"],
    index=color_map.keys()
)

# L√†m tr√≤n gi√° tr·ªã ƒë·∫øn 2 ch·ªØ s·ªë th·∫≠p ph√¢n
description_df.iloc[:,0] = description_df.iloc[:,0].apply(
    lambda x: float("{:.2f}".format(x))
)

# Th√™m t√™n cho ch·ªâ m·ª•c
description_df.index.name = "Lo√†i hoa"

# Hi·ªÉn th·ªã b·∫£ng m√¥ t·∫£
description_df.style.background_gradient(cmap="Blues", vmin=-1, vmax=1)
```

Nh·∫≠n x√©t:

- Nh√¨n chung, ·ªü c·∫£ ba lo√†i hoa, ta th·∫•y chi·ªÅu d√†i v√† chi·ªÅu r·ªông c√°nh hoa c√≥ m·ªëi quan h·ªá tuy·∫øn t√≠nh theo chi·ªÅu d∆∞∆°ng (t∆∞∆°ng quan thu·∫≠n). - Trong ƒë√≥, quan h·ªá tuy·∫øn t√≠nh gi·ªØa chi·ªÅu d√†i v√† chi·ªÅu r·ªông c√°nh hoa ·ªü lo√†i "Iris-versicolor" c√≥ m·ª©c ƒë·ªô t∆∞∆°ng quan r·∫•t m·∫°nh (0.79).   - Ng∆∞·ª£c l·∫°i, c√°c c√°nh hoa ·ªü lo√†i "Iris-virginica" th∆∞·ªùng d√†i v√† r·ªông h∆°n kh√° nhi·ªÅu so v·ªõi c√°nh hoa ·ªü hai lo√†i c√≤n l·∫°i.
  - Trong khi ƒë√≥, lo√†i "Iris-versicolor" th∆∞·ªùng c√≥ k√≠ch th∆∞·ªõc c√°nh hoa ·ªü m·ª©c trung b√¨nh.
