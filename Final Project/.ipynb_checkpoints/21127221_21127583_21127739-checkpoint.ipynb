#### 📌 Kiểu dữ liệu hiện tại của mỗi cột là gì? Có cột nào có kiểu dữ liệu không phù hợp không?

```python
data.info()
```

- `is_canceled` là cờ lệnh để thể hiện liệu đặt phòng có bị huỷ hay không, do đó cần chuyển về object để phân vào categorical.
- `children` chỉ số lượng trẻ em, do đó cần chuyển về dạng int64. Tuy nhiên vì dữ liệu đang chứa NULL, do đó sẽ chuyển về dạng int64 sau khi fill missing value.
- `is_repeated_guest` là cờ lệnh để thể hiện khách hàng có từng lưu trú ở đó không, do đó cần chuyển về object để phân vào categorical.
- `reservation_status_date` biểu diễn cho ngày cập nhật trạng thái đặt phòng cuối cùng, do đó cần chuyển về dạng datetime.

```python
data['is_canceled'] = data['is_canceled'].astype(str)
data['is_repeated_guest'] = data['is_repeated_guest'].astype(str)
data["reservation_status_date"] = pd.to_datetime(data["reservation_status_date"])
```

#### 📌 Đối với mỗi cột có kiểu dữ liệu số, các giá trị được phân bố như thế nào?

📚 Đối với các cột có kiểu dữ liệu số, nhóm sẽ tính toán một số thống kê cơ bản:
- Tỷ lệ phần trăm (từ 0 đến 100) giá trị bị thiếu
- The min
- The lower quartile (phân vị 25)
- The median (phân vị 50)
- The upper quartile (phân vị 75)
- The max


```python
num_col_info_df = data.select_dtypes(exclude=['object', 'bool'])

def missing_ratio(s):
    return (s.isna().mean() * 100)

def median(df):
    return (df.quantile(0.5))

def lower_quartile(df):
    return (df.quantile(0.25))

def upper_quartile(df):
    return (df.quantile(0.75))

num_col_info_df = num_col_info_df.agg([missing_ratio, "min", lower_quartile, median, upper_quartile, "max"])
num_col_info_df.T
```

- Cột `company` có tỉ lệ thiếu rất cao, lên đến hơn 90% (94.306893%), do đó nhóm sẽ tiến hành loại bỏ cột này.
- Cột `children` với tỉ lệ thiếu nhỏ, nhóm điền bằng giá trị mode của thuộc tính này.
- Cột `agent` với tỉ lệ thiếu khá cao, nhóm điền bằng giá trị median của thuộc tính này do giá trị của nó có phân bố rộng.

```python
data.drop('company', axis=1, inplace=True)
data['children'].fillna(data['children'].mode()[0], inplace=True)
data['agent'].fillna(data['agent'].median(), inplace=True)
data['children'] = data['children'].astype(np.int64)
```

```python
# TEST
data.select_dtypes(exclude=['object', 'bool']).agg([missing_ratio, "min", lower_quartile, median, upper_quartile, "max"]).T
```

#### 📌 Đối với mỗi cột có kiểu dữ liệu không phải là số, các giá trị được phân phối như thế nào?

📚 Đối với các cột có kiểu dữ liệu không phải là số, nhóm sẽ tính toán:
- Tỷ lệ phần trăm (từ 0 đến 100) giá trị bị thiếu
- Số lượng giá trị khác nhau
- Hiển thị một vài giá trị và phần trăm (từ 0 đến 100) của từng giá trị


```python
cat_col_info_df = data.select_dtypes(include=['object', 'bool'])

def missing_ratio(s):
    return (s.isna().mean() * 100)

def num_values(s):
    return len(s.value_counts())

def value_ratios(s):
    totalCount = (~s.isna()).sum()
    return ((s.value_counts()/totalCount*100).round(1)).to_dict()

cat_col_info_df = cat_col_info_df.agg([missing_ratio, num_values, value_ratios])
cat_col_info_df.T
```

Cột `country` với tỉ lệ thiếu là 0.517186% , không đáng kể, do đó nhóm sẽ  điền giá trị thiếu bằng giá trị mode của cột.

```python
data['country'].fillna(data['country'].mode()[0], inplace=True)
```

```python
# TEST
data.select_dtypes(include=['object', 'bool'])\
    .agg([missing_ratio, num_values, value_ratios]).T
```

- Mối tương quan nghịch lớn nhất là -0.51 giữa thuộc tính `arrival_date_year` và `arrival_date_week_number`.
- Ngoài ra còn một số mối quan hệ  giữa các biến:
    - `previous_bookings_not_canceled` - `previous_cancellations`: 0.39
    - `adr` - `children`: 0.33
    - `stays_in_week_nights` - `lead_time`: 0.31
- Hầu hết các giá trị tương quan của các thuộc tính còn lại đều xấp xỉ bằng 0, chứng tỏ các thuộc tính này không có mối quan hệ chặt chẽ với nhau.

Khi đã có cái nhìn tổng quát về mối quan hệ giữa các thuộc tính, chúng ta sẽ dùng Scatter plot để phân tích và hiểu rõ hơn mối quan hệ của 2 thuộc tính `stays_in_weekend_nights` - `stays_in_week_nights`

#### Sử dụng Scatter plot để phân tích dữ liệu 2D: 

##### Mối quan hệ của stays_in_week_nights và stays_in_weekend_nights

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax = sns.scatterplot(data=data, x="stays_in_week_nights", y="stays_in_weekend_nights")
ax.set_aspect('equal', adjustable='box')
plt.xlabel('stays_in_week_nights', size = 15)
plt.ylabel('stays_in_weekend_nights', size = 15)
plt.locator_params(axis='x', integer=True)
plt.locator_params(axis='y', integer=True)
plt.xticks(range(0, max(data['stays_in_week_nights'])+1, 2))
plt.yticks(range(0, max(data['stays_in_weekend_nights'])+1, 2))

x_min = min(data["stays_in_week_nights"])
y_min = min(data["stays_in_weekend_nights"])
x_max = max(data["stays_in_week_nights"])
y_max = max(data["stays_in_weekend_nights"])

plt.grid(True, zorder=1, alpha=0.5)
plt.gca().set_axisbelow(True)
plt.plot([x_min, x_max], [y_min, y_max], color='red', linestyle='--')
plt.show()
```

- Giá trị của `stays_in_week_nights` phần lớn tập trung trong khoảng từ 0 đến 22.
- Giá trị của `stays_in_weekend_nights` phần lớn tập trung trong khoảng từ 0 đến 10.
- Chúng ta sẽ vẽ một đường thẳng đi qua điểm nhỏ nhất và lớn nhất trên biểu đồ để dễ quan sát. Qua quan sát, có vẻ như hai thuộc tính này có mối quan hệ rất chặt chẽ với nhau khi các điểm dữ liệu hầu hết tập trung xung quanh đường thẳng và không có điểm outlier.
- Tuy nhiên, để chắc chắn hơn, chúng ta sẽ tiến hành tìm đường thẳng tuyến tính cho nó.

```python
fig, ax = plt.subplots(figsize=(15, 10))
size_of_data = data.groupby(['stays_in_week_nights', 'stays_in_weekend_nights']).size().reset_index(name='size')
sns.scatterplot(data=size_of_data, x="stays_in_week_nights", y="stays_in_weekend_nights", size='size', sizes=(20, 500), ax=ax)

x = data["stays_in_week_nights"]
y = data["stays_in_weekend_nights"]
coefficients = np.polyfit(x, y, 1)
polynomial = np.poly1d(coefficients)
y_fit = polynomial(x)
plt.plot(x, y_fit, color='pink', linestyle='--', label=f'Linear Trend: y = {coefficients[0]:.2f}x + {coefficients[1]:.2f}')

ax.set_aspect('equal', adjustable='box')
plt.xlabel('stays_in_week_nights', size = 15)
plt.ylabel('stays_in_weekend_nights', size = 15)
plt.locator_params(axis='x', integer=True)
plt.locator_params(axis='y', integer=True)
plt.grid(True, zorder=1, alpha=0.5)
plt.gca().set_axisbelow(True)
plt.xticks(range(0, max(data['stays_in_week_nights'])+1, 2))
plt.yticks(range(0, max(data['stays_in_weekend_nights'])+1, 2))
plt.show()
```

Lần này, chúng ta sẽ thêm vào biểu đồ số lượng của các điểm dữ liệu được biểu thị thông qua độ lớn của chấm tròn. Và tạo một đường thẳng hồi quy cho tập dữ liệu của 2 thuộc tính `stays_in_weekend_nights` - `stays_in_week_nights`, trực quan lên biểu đồ.
Một số nhận xét về biểu đồ mới:
- Các điểm dữ liệu không còn thực sự khớp với đuòng thẳng hồi quy mới.
- Hầu hết các điểm dữ liệu tập trung trong khoảng: 0 <= `stays_in_week_nights` <= 6; 0 <= `stays_in_weekend_nights` <= 3.
- Do số lượng các điểm dữ liệu phân bố không đều, nên đường thẳng chỉ khớp ở các dữ liệu có giá trị nhỏ; càng lớn, dữ liệu càng không khớp với đường thẳng.
- Có thể thấy, `stays_in_weekend_nights` - `stays_in_week_nights` có mối quan hệ đồng biến nhưng không hoàn toàn.
- Khách hàng có xu hướng chọn dịch vụ ở qua đêm vào các ngày trong tuần nhiều hơn các ngày cuối tuần.
- Khi số đêm ở lại thuộc các ngày trong tuần tăng thì số đêm ở lại thuộc các ngày cuối tuần cũng tăng, nhưng không tăng nhiều bằng.

#### Tính tỷ trọng đối với hai biến cate

##### Tỷ trọng loại phòng được đặt tại các loại khách sạn khác nhau

```python
city_hotel_counts = data[data['hotel'] == 'City Hotel']['reserved_room_type'].value_counts()
resort_hotel_counts = data[data['hotel'] == 'Resort Hotel']['reserved_room_type'].value_counts()

city_hotel_percentages = city_hotel_counts / city_hotel_counts.sum() * 100
resort_hotel_percentages = resort_hotel_counts / resort_hotel_counts.sum() * 100

threshold = 2
city_hotel_percentages_others = city_hotel_percentages[city_hotel_percentages >= threshold]
city_hotel_percentages_others['Khác'] = city_hotel_percentages[city_hotel_percentages < threshold].sum()

resort_hotel_percentages_others = resort_hotel_percentages[resort_hotel_percentages >= threshold]
resort_hotel_percentages_others['Khác'] = resort_hotel_percentages[resort_hotel_percentages < threshold].sum()

plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
plt.pie(city_hotel_percentages_others, autopct='%1.1f%%')
plt.title('City Hotel')
plt.legend(labels=city_hotel_percentages_others.index)

plt.subplot(1, 2, 2)
plt.pie(resort_hotel_percentages_others, autopct='%1.1f%%')
plt.title('Resort Hotel')
plt.legend(labels=resort_hotel_percentages_others.index)

plt.suptitle("Tỷ trọng loại phòng được đặt tại các loại khách sạn")
plt.tight_layout()
plt.show()
```

Nhận xét:
- Nhìn chung, tại các loại khách sạn, 2 loại phòng được yêu thích nhất lần lượt là loại A, loại D.
- Trong khi tại `City hotel` thì phòng loại A chiếm tỉ lệ rất lớn 71%, phòng loại D chiếm 20% và 9% cho các loại còn lại. Thì bên `Resort hotel` tỉ lệ giữa các loại phòng căn bằng hơn, loại A vẫn nhiều nhất nhưng chỉ chiếm 55%, loại D 19.5%, loại E 13.5% và 12% cho các loại còn lại.

##### Tỷ trọng hủy đặt phòng ở các nhóm khách hàng

```python
name_customer_type = data['customer_type'].value_counts().index

customer_type = pd.DataFrame()
for i in name_customer_type:
    customer_type[i] = data[data['customer_type'] == i]['is_canceled'].value_counts()

plt.figure(figsize=(15, 10))
for i in range(name_customer_type.size):
    plt.subplot(2, 2, i+1)
    plt.pie(list(customer_type[name_customer_type[i]]), autopct='%1.1f%%')
    plt.title(name_customer_type[i])
    plt.legend(labels=["no canceled", "canceled"])

plt.suptitle("Tỷ trọng hủy đặt phòng ở các nhóm khách hàng")
plt.tight_layout()
plt.show()
```

Nhận xét:
- Tỉ lệ hủy phòng ở nhóm khách hàng `transient` là cao nhất với khoảng 30%, do đặc thù của nhóm khách hàng này là mọi thứ không có sự sắp xếp trước nên tỉ lệ hủy phòng cao.
- Tỉ lệ hủy phòng ở nhóm khách hàng `group` là thấp nhất với khoảng 10%, do đặc thù của nhóm khách hàng này là có sự sắp xếp trước và có sự tham gia của nhiều người nên tỉ lệ hủy phòng thấp.
- Ở nhóm `transient-party` và `contract` là xấp xỉ nhau, khoảng 15-16%. Tỉ lệ hủy phòng không quá cao.
- Các khách sạn cần có quy định phù hợp cũng như tạo điều kiện thuận lợi cho khách hàng để đảm bảo quyền lợi của khách sạn mà vẫn đảm bảo sự hài lòng của khách hàng khi hủy đặt phòng ở các nhóm khách hàng có tỉ lệ hủy phòng cao.
